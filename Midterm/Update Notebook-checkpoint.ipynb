Wow, lots of detail! Personally I'm not interested in the brand of medicine so I will sum these columns to trim the features. I'll also drop the provider's NPI, since there's not much information to glean from that.

```python
df['total_prescribed'] = df.iloc[:, 5:255].sum(axis=1)
df_final = df.drop(df.columns[5:255], axis=1)
df_final = df_final.drop(['NPI'], axis=1)
df_final.head()
```

Much better! The first plot I had in mind was a bar plot by state. I'll do a quick sum.

```python
dims = (20, 15)
fig, ax = plt.subplots(figsize=dims)
sns.barplot(data=df_final, x='State',y='total_prescribed', estimator=sum)
```

Next, I will create a bar plot by specialization. The interaction in this plot will be an explosion if the mouse hovers over. For the purpose of demonstration, I am hardcoding this, but in the actual program it will move.

```python
dims = (20, 15)
fig, ax = plt.subplots(figsize=dims)
sums = df_final.total_prescribed.groupby(df_final.Specialty).sum()
explode = np.zeros(len(sums))
explode[24] = .1
ax.axis('equal');
plt.pie(sums, labels=sums.index,explode=explode,autopct='%1.1f%%',
        shadow=True, startangle=90);
plt.tight_layout()
plt.show()
```
